```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV
df = pd.read_csv("loss_pretrain.csv")

# 提取数据
epoch = df["epoch"][:100]
train_loss = df["train_loss"][:100]
val_loss = df["val_loss"][:100]

# 画图
plt.figure(figsize=(10,5))

plt.plot(epoch, train_loss, label="Train Loss",linewidth=2)
plt.plot(epoch, val_loss, label="Validation Loss",linewidth=2)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.ylim(0, 0.2)
plt.title("Pretraining Loss Curve")

plt.legend()
# plt.grid(True)

plt.tight_layout()
# plt.show()

plt.savefig("loss_pretrain1.png")
```
