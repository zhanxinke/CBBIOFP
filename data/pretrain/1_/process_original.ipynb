```python
import pandas as pd
from pathlib import Path
import logging

df_Thera = pd.read_csv("main_fileter.csv")
df_Thera.head()

```

```python
df_Thera = df_Thera[['Function','Sequence']]
df_Thera
```

```python
def filter_peptides(df):
    # 计算序列长度
    df['length'] = df['Sequence'].str.len()
    
    # 条件1: 长度在 [5, 40] 之间
    cond_length = df['length'].between(5, 50)
    
    # 条件2: X 字符比例 < 1/3
    cond_x = df['Sequence'].str.count("X") / df['length'] < (1/3)
    
    # 结合条件
    df_filtered = df[cond_length & cond_x].copy()
    
    return df_filtered.drop(columns=['length'])
# df_filtered.drop(columns=['length'])
```

```python
df_clean = filter_peptides(df_Thera)
# df_clean['function'] = df_clean['function'].replace("Growth_regeneration", "Growth_Regeneration")
df_clean.drop_duplicates(inplace=True)
df_clean.reset_index(drop=True)
# df_clean.to_csv('therapeptides_clean.csv', index=False)
```

```python
df_clean['Sequence'] = df_clean['Sequence'].str.upper()
df_clean = df_clean[df_clean['Sequence'].str.match(r'^[A-Za-z]+$')]
df_clean.reset_index(drop=True)
```

```python
df_clean.drop_duplicates(inplace=True)
df_clean.reset_index(drop=True)
df_clean.to_csv('therapeptides_clean.csv', index=False)
```
